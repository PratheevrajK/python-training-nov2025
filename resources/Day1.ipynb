## Console Input/Output
#### input() method/function-to take input from user

```python
# take input for principal amount and calculate simple interest
#typecast input to int
principal_amount=int(input("Enter the principal amount : "))
print(principal_amount)
print(type(principal_amount))
```

```python
#calculate simple interest
RATE_OF_INTEREST=10.5
simple_interest=principal_amount*RATE_OF_INTEREST/100 
print(simple_interest)
```

```python
print("principal amount: ",principal_amount)
print("Rate of interest: ",RATE_OF_INTEREST)
print("simple interest: ",simple_interest)
```

```python
# take input for principal amount and calculate simple interest
#typecast input to float type
principal_amount=float(input("Enter the principal amount : "))
print(principal_amount)
print(type(principal_amount))
```
