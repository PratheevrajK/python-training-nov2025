## Arithmetic Operators
#### Operator	Meaning	Example	Output
####  +	Addition	3 + 2	5
#### -	Subtraction	5 - 3	2
#### *	Multiplication	4 * 3	12
#### /	Floating division	7 / 2	3.5
#### //	Floor division	7 // 2	3
#### %	Modulus (remainder)	7 % 2	1
#### **	Exponent	2 ** 3	8


```python
#Take input for two numbers
num1=int(input("Enter number1"))
num2=int(input("Enter number2"))
```

```python
#Addition
sum=num1+num2
print("Sum of {} and {} is {}".format(num1,num2,sum))
```

```python
#Subtraction
sub=num1-num2
print("Difference between {} and {} is {}".format(num1,num2,sub))
```

```python
#Multiplication
mul=num1*num2
print("Product of {} and {} is {}".format(num1,num2,mul))
```

```python
#Division
div=num1/num2
print("If {} divide by {} then result is {}".format(num1,num2,div))
```

```python
#Floor Division
floor_div=num1//num2
print("If {} divide by {} then result is {}".format(num1,num2,floor_div))
```

```python
#Remainder
rem=num1%num2
print("If {} divide by {} then remainder is {}".format(num1,num2,rem))
```

```python
#Exponent
exp=num1**num2
print("{} ^ {} is {}".format(num1,num2,exp))
```

```python
sum+=5
print("Updated sum=", sum)
```

```python
sum-=5
print("Updated sum=", sum)
```

## Logical or Boolean operators
#### Operator	Meaning	Example	Output
#### and	True if both are True	True and False	False
#### or	True if at least one is True	True or False	True
#### not	Reverses Boolean value	not True	False


```python
a=10
b=20
c=30
res=(a<b)and(b<c)
print(res)
```

```python
res=(a<b)or(b<c)
print(res)
```

```python
res=not(a<b)
print(res)
```

#### if...elif....else statement

```python
# WAP to display grade of a student
""" score>=90 then Grade A
    score>=70 then Grade B
    score>=50 then Grade C
    else Grade D
    """
score=int(input("Enter score"))
if score>=90:
    print("Grade A")
elif score>=70:
    print("Grade B")
elif score>=50:
    print("Score C")
else:
    print("Score D")
```

```python
# Ask the user for the amount in USD
usd_input = input("Enter the amount in USD: ")

# Convert input to float
usd = float(usd_input)

# Constants (exchange rates)
USD_TO_EUR = 0.93
USD_TO_JPY = 151.78

# Processing
eur = usd * USD_TO_EUR
jpy = usd * USD_TO_JPY

# Output
print(f"USD Amount: {usd}")
print(f"Equivalent in EUR: {eur}")
print(f"Equivalent in JPY: {jpy}")
```

```python
# Ask the user for resistor values
r1_input = input("Enter the value of resistor R1 (in ohms): ")
r2_input = input("Enter the value of resistor R2 (in ohms): ")

# Convert inputs to float
R1 = float(r1_input)
R2 = float(r2_input)

# Calculations
R_series = R1 + R2
R_parallel = (R1 * R2) / (R1 + R2)

# Output
print(f"Series Resistance (R_series): {R_series} ")
print(f"Parallel Resistance (R_parallel): {R_parallel} ")
```

```python
# Ask the user for weight and height
weight_input = input("Enter your weight in kilograms (kg): ")
height_input = input("Enter your height in meters (m): ")

# Convert inputs to float
weight = float(weight_input)
height = float(height_input)

# Calculate BMI
bmi = weight / (height ** 2)

# Output the BMI
print(f"Your BMI is: {bmi:.2f}")
```
